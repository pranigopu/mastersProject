# REGRESSION PROBLEM

# Setup

```python
import numpy as np
import matplotlib.pyplot as plt

# Small helper function to add details to plots:
def plot(title, xlabel, ylabel, legend=False):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if legend: plt.legend()
    plt.show()
```

# Generating synthetic data

```python
x = np.arange(-20, 20)
y = (x**2)*np.sin(x)
plt.plot(x, y)
plot('Generated synthetic data', 'x', 'y')
```
